```python
from fastkan import FastKANLayer

d_in = 2
d_out = 3

layer = FastKANLayer(
    d_in, d_out,
    use_base_update=False,
    use_layernorm=False
)

x, y = layer.plot_curve(0, 1, num_pts=1000, num_extrapolate_bins=3)
x.shape, y.shape
```

```python
import matplotlib.pyplot as plt

for i in range(d_in):
    for j in range(d_out):
        x, y = layer.plot_curve(i, j, 200, num_extrapolate_bins=3)
        plt.plot(x.numpy(), y.numpy(), label=f"phi {i, j}")
plt.legend()
```
